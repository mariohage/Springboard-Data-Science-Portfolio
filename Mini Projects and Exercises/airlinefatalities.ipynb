```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("airlinedata.csv")
```

```python
df.shape
```

```python
df.describe()
## sorting for 2000-2014
df = df.sort_values('fatalities_00_14', ascending=False)

top20 = df.head(20)

display(top20)
```

```python
top20.plot.barh(x='airline', y='fatalities_00_14', sort_columns=True)
```

```python
## sorting for 1985-1999


df2 = df.sort_values('fatalities_85_99', ascending=False)

top20v2 = df2.head(20)

display(top20v2)
```

```python
top20v2.plot.barh(x='airline', y='fatalities_85_99', sort_columns=True)
```

```python
## Improvement over the years (Based on change in fatalities between the two timeperiods 1985-1999 and 2000-2014)

df['fatalitiesdiff'] = df['fatalities_00_14'] - df['fatalities_85_99']
```

```python

df3 = df.sort_values('fatalitiesdiff', ascending=False)

top20v3 = df3.head(20)

display(top20v3)
```

```python
top20v3.plot.barh(x='airline', y='fatalitiesdiff', sort_columns=True)
```
